```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    self.head=Node(a[0],None,None)
    temp=self.head
    for i in range(1,len(a)):
      n=Node(a[i],None,temp)
      temp.next=n
      temp=n
      if i==len(a)-1:
        n.next=self.head
        self.head.prev=n
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    temp=self.head
    count=1
    while temp.next!=self.head:
      temp=temp.next
      count+=1
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    temp=self.head
    while temp.next!=self.head:
      print(temp.element,end=",")
      temp=temp.next
    print(self.head.prev.element)

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    temp=self.head.prev
    while temp!=self.head:
      print(temp.element,end=",")
      temp=temp.prev
    print(self.head.element)

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    temp=self.head
    idx1=0
    while temp.next!=self.head:
      if idx1==idx:
        return temp
      temp=temp.next
      idx1+=1
      if temp.next==self.head:
        if idx1==idx:
          return temp
    return None
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    temp=self.head
    idx=0
    while temp.next!=self.head:
      if temp.element==elem:
        return idx
      temp=temp.next
      idx+=1
      if temp.next==self.head:
        if temp.element==elem:
          return idx
    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    if idx<0 or idx>self.countNode():
      return "Invalid Index"
    else:
      n=Node(elem,None,None)
      if idx==0:
        n.next=self.head
        n.prev=self.head.prev
        self.head.prev.next=n
        self.head.prev=n
        self.head=n
      #elif idx==self.countNode():
        #n.prev=self.head.prev
        #n.next=self.head
        #self.head.prev.next=n
        #self.head.prev=n

      else:
        temp=self.head
        idx1=0
        while temp.next!=self.head:
          if idx1==idx-1:
            n.prev=temp
            n.next=temp.next
            temp.next=n
            n.next.prev=n
          temp=temp.next
          idx1+=1
          if temp.next==self.head and idx1==idx-1:
            n.prev=temp
            n.next=temp.next
            temp.next=n
            n.next.prev=n
            break

  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    if idx==0:
      self.head.prev.next=self.head.next
      self.head.next.prev=self.head.prev
      temp=self.head.next
      self.head.prev=None
      self.head.next=None
      elem=self.head.element
      self.head=temp
      return str(elem)
    else:
      temp=self.head
      idx1=0
      while temp.next!=self.head:
        if idx1==idx-1:
          #1-2-3
          elem=temp.next.element
          temp1=temp.next
          temp.next=temp.next.next
          temp.next.prev=temp
          
          return str(elem)
        temp=temp.next
        idx1+=1
```
